```python
import pandas as pd
import matplotlib.pyplot as plt

weather_list = pd.read_csv('./parsed_weather_data.csv')
weather_list['Day Temperature'] = pd.to_numeric(weather_list['Day Temperature'].str.replace('−', '-'), errors='coerce')
weather_list['Night Temperature'] = pd.to_numeric(weather_list['Night Temperature'].str.replace('−', '-'), errors='coerce')


overall_max_temp = max(weather_list[['Day Temperature', 'Night Temperature']].values.flatten())
overall_min_temp = min(weather_list[['Day Temperature', 'Night Temperature']].values.flatten())


weather_list.head(pd.get_option('display.max_rows'))
```

```python
weather_list.describe()

days = weather_list['Day']
day_temps = weather_list['Day Temperature']


plt.figure(figsize=(12, 12))

plt.subplot(2, 1, 1)
plt.plot(days, day_temps, label='Day Temperature', marker='o', color='yellow')


for day, temp in zip(days, day_temps):
    plt.text(day, temp, str(temp), ha='center', va='bottom', fontsize=13, color='black')


plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Day Temperatures')

plt.yticks(range(overall_min_temp + -3, overall_max_temp + 3))
plt.xticks(range(0, int(days.max()) + 1))

plt.legend()
```

```python
weather_list.describe()

days = weather_list['Day']
night_temps = weather_list['Night Temperature']

plt.figure(figsize=(12, 12))

# Night Temperature Plot
plt.subplot(2, 1, 1)
plt.plot(days, night_temps, label='Night Temperature', marker='o', color='blue')


for day, temp in zip(days, night_temps):
    plt.text(day, temp, str(temp), ha='center', va='bottom', fontsize=13, color='black')

plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Night Temperatures')

plt.yticks(range(overall_min_temp + -3, overall_max_temp + 3))
plt.xticks(range(0, int(days.max()) + 1))

plt.legend()
```

```python
all_temperatures = pd.concat([weather_list['Day Temperature'], weather_list['Night Temperature']])
overall_average_temp = all_temperatures.mean()

print("Overall average temp:", overall_average_temp)

print("Warmest temp:", overall_max_temp)

print("Coldest temp:", overall_min_temp)
```
